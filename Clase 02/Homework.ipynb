1) Abrir el archivo "Emisiones_CO2.csv" y cargar sus datos en un diccionario.

```python
import os
archivo = open('Emisiones_CO2.csv','r',encoding='ANSI')
```

```python
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
```

```python
next(archivo)
for line in archivo:
    campos = line.split(sep='|')
    dicc_emisiones['cod_pais'].append(campos[0])
    dicc_emisiones['nom_pais'].append(campos[1])
    dicc_emisiones['region'].append(campos[2])
    dicc_emisiones['anio'].append(campos[3])
    dicc_emisiones['co2'].append(campos[4])
    dicc_emisiones['co2_percapita'].append(campos[5])


```

e) ¿Cuál es el total de emisiones de CO2 para 'América Latina y Caribe' en el año 2010?

```python
# Cambiando los valores a float de la key co2
for z,it in enumerate(dicc_emisiones['co2']):
    
    it = it.replace('.','')
    it = it.replace(',','.')
    if it == '':
        it = 0
    dicc_emisiones['co2'][z] = float(it)

print(dicc_emisiones['co2'])
```

```python
# sumar valores de america latina
suma_emisiones=0
for w,val in enumerate(dicc_emisiones['co2']):
    if(dicc_emisiones['region'][w] == 'América Latina y Caribe' and dicc_emisiones['anio'][w]=='2010'):
        suma_emisiones += val

print(f'La suma de emisones del 2010 en ALyC es {suma_emisiones}')
```
